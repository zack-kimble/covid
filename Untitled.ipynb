```python
import requests
import pandas as pd
import io
import seaborn as sns
```

```python
r = requests.get('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Deaths.csv')
deaths_df = pd.read_csv(io.StringIO(r.text))
id_cols = ['Province/State', 'Country/Region', 'Lat', 'Long']
deaths_df = deaths_df.melt(id_vars = id_cols, var_name = 'date',value_name = 'deaths')
```

```python
countries_with_deaths = deaths_df[deaths_df.deaths > 0]['Country/Region'].unique()
countries_with_outbreak = deaths_df[deaths_df.deaths >= 10]['Country/Region'].unique()
deaths_df = deaths_df[deaths_df['Country/Region'].apply(lambda x: x in countries_with_deaths)]
outbreak_df = deaths_df[deaths_df['Country/Region'].apply(lambda x: x in countries_with_outbreak)]
```

```python
sns.set(rc={'figure.figsize':(20,8.27)})
sns.lineplot(data = outbreak_df, x = 'date', y = 'deaths', hue='Country/Region') 
```
